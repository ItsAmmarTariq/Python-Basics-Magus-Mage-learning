```python


df = pd.DataFrame({'City':['New York (City)', 'Parague', 'New Delhi (Delhi)', 'Venice', 'new Orleans'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy', 'Tech_Summit'],
                    'Cost':[10000, 5000, 15000, 2000, 12000]})

# city column has a extra city brackets remove it using the regex 

import re


def Clean_names(City_name):
    # Search for opening bracket in the name followed by
    # any characters repeated any number of times
    if re.search('\(.*', City_name):
 
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', City_name).start()
 
        # return the cleaned name
        return City_name[:pos]
    else:
        return City_name
 
```

```python
print(df)
df['City']=df['City'].apply(Clean_names)
df
```

```python
import pandas as pd

# Read data from the CSV file
df = pd.read_csv('facebook_data.csv', parse_dates=['timestamp'])

# Filter rows where 'action' is either 'page_load' or 'scroll_down'
if_sessions = df[df['action'].isin(['page_load', 'scroll_down'])]

# Calculate session time for each user and date
if_sessions['session_time'] = if_sessions.groupby(['user_id', if_sessions['timestamp'].dt.date])['timestamp'].transform(lambda x: x.max() - x.min())

# Calculate average session time for each user
average_session_time = if_sessions.groupby('user_id')['session_time'].mean().reset_index()

# Display the result
print(average_session_time)
```
