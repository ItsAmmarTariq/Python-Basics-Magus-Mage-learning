### Dealing with missing values

```python
import pandas as pd
data=pd.read_csv('temp_data.csv')
data.head()
```

```python
data.dropna()
data
```

```python
data.dropna()
# this operation is along the x-axis 
```

```python
data.dropna(axis=1)
# here it deals with vaules along column wise 
```

```python
data.dropna(how='any')
```

```python
data.dropna(how='all')
```

```python
data.dropna(subset='Temperature_May')
data.head()
```

```python
data.dropna(inplace=True)
data
# return a new dataset with no NAN values
```

```python
data.dropna(thresh=1)
# delete one row only having NAN values 
```

```python
data.fillna('Python')
# fill NAN values as Python 
# can be any value
```

```python
# for  a particular values in a particular value 
# i can use 
data.fillna({'Temperature_Jan':'jan','Temperature_Feb':'ffeb'})
# rest of the values reian NAN
```

```python
data.fillna(method='ffill')
```

```python
data.fillna(method='bfill',axis=1)
```

```python
var=data.replace(32.5,'python')
var
```

```python
# in replace we can also use regex
var.replace('[a_z]','pytho1',regex=True)
# also we can specify a particular column with a particular value
var.replace({'Temperature_Jan':'[a-z]'},22,regex=True)
```
